```python
import os
from ultralytics import YOLO

# 1. ì €ì¥ ê²½ë¡œ ì„¤ì • (ë¡œì»¬ í™˜ê²½ì— ë§ê²Œ ìˆ˜ì •)
save_dir = '/content/drive/MyDrive/yolol8_checkpoints'
os.makedirs(save_dir, exist_ok=True)

# 2. data.yaml ë‚´ìš© (í•™ìŠµ/ê²€ì¦ ì´ë¯¸ì§€ ê²½ë¡œë¥¼ ë¡œì»¬ ê²½ë¡œë¡œ ë³€ê²½)
yaml_content = """
train: /content/drive/MyDrive/Unzipped/yolo/images/train
val: /content/drive/MyDrive/Unzipped/yolo/images/val

nc: 4
names: ['ì›í”¼ìŠ¤', 'ìƒì˜', 'í•˜ì˜', 'ì•„ìš°í„°']
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("âœ… data.yaml ìƒì„± ì™„ë£Œ")

# 3. ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (ë¡œì»¬ best.pt ë˜ëŠ” last.pt ê²½ë¡œë¡œ ë³€ê²½)
model = YOLO('yolov8l.pt')

# 4. í•™ìŠµ ì‹œì‘ (ì¤‘ê°„ ì €ì¥ í¬í•¨, ì´ì–´ì„œ í•™ìŠµ)
model.train(
    data='data.yaml',
    epochs=50,
    imgsz=640,
    batch=8,
    project=save_dir,
    name='exp',
    exist_ok=True,
    save=True,
    save_period=1,
    # resume=True
)
```

```python
import os
import cv2
from ultralytics import YOLO

# 1. Google Drive ë§ˆìš´íŠ¸
from google.colab import drive
drive.mount('/content/drive')

# 2. YOLOv8 ëª¨ë¸ ë¡œë“œ (Google Drive ê²½ë¡œì— best.pt ìˆë‹¤ê³  ê°€ì •)
model = YOLO('/content/drive/MyDrive/yolol8_checkpoints/exp/weights/best.pt')

# 3. í¬ë¡­ ì´ë¯¸ì§€ ì €ì¥ í•¨ìˆ˜
def yolo_detect_and_crop(image_path, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    img = cv2.imread(image_path)
    results = model(img, conf=0.1)  # confidence threshold ì¡°ì • ê°€ëŠ¥

    if len(results[0].boxes) == 0:
        print(f"â— íƒì§€ ì‹¤íŒ¨: {image_path}")
        return

    base_name = os.path.splitext(os.path.basename(image_path))[0]

    for i, box in enumerate(results[0].boxes.xyxy):
        x1, y1, x2, y2 = map(int, box.tolist())
        crop = img[y1:y2, x1:x2]
        crop_path = os.path.join(save_dir, f"{base_name}_crop_{i}.jpg")
        cv2.imwrite(crop_path, crop)
        print(f"âœ… í¬ë¡­ ì €ì¥: {crop_path}")

# 4. í´ë” ë‚´ ì´ë¯¸ì§€ ì¼ê´„ ì²˜ë¦¬ í•¨ìˆ˜
def yolo_detect_and_crop_folder(input_folder, save_dir):
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(input_folder, filename)
            yolo_detect_and_crop(image_path, save_dir)

    print(f"âœ… ì „ì²´ í´ë” ì²˜ë¦¬ ì™„ë£Œ. í¬ë¡­ ì´ë¯¸ì§€ ì €ì¥ ìœ„ì¹˜: {save_dir}")

# 5. ì‚¬ìš© ì˜ˆì‹œ (Google Drive ê²½ë¡œì— ë§ê²Œ ìˆ˜ì •)
input_folder = '/content/drive/MyDrive/Unzipped/yolo/images'
save_crop_folder = '/content/drive/MyDrive/crop'

yolo_detect_and_crop_folder(input_folder, save_crop_folder)
```

```python
import os
import cv2

def yolo_detect_and_crop_debug(image_path, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    img = cv2.imread(image_path)

    if img is None:
        print(f"âŒ ì´ë¯¸ì§€ ì½ê¸° ì‹¤íŒ¨: {image_path}")
        return

    results = model(img, conf=0.1)
    num_boxes = len(results[0].boxes)
    print(f"â„¹ï¸ {image_path} â†’ íƒì§€ ë°•ìŠ¤ ê°œìˆ˜: {num_boxes}")

    if num_boxes == 0:
        print(f"âš ï¸ íƒì§€ ê²°ê³¼ ì—†ìŒ â†’ í¬ë¡­ ë¶ˆê°€")
        return

    base_name = os.path.splitext(os.path.basename(image_path))[0]

    for i, box in enumerate(results[0].boxes.xyxy):
        try:
            x1, y1, x2, y2 = map(int, box.tolist())
            # ì¢Œí‘œ ê²½ê³„ ì²´í¬ ë° ë³´ì •
            x1, y1 = max(0, x1), max(0, y1)
            x2, y2 = min(img.shape[1], x2), min(img.shape[0], y2)

            if x2 <= x1 or y2 <= y1:
                print(f"âš ï¸ ì˜ëª»ëœ ì¢Œí‘œ: x1={x1}, y1={y1}, x2={x2}, y2={y2} (ë¬´ì‹œ)")
                continue

            crop = img[y1:y2, x1:x2]
            if crop.size == 0:
                print(f"âš ï¸ í¬ë¡­ ì˜ì—­ì´ ë¹„ì–´ìˆìŒ (ë¬´ì‹œ): {x1}, {y1}, {x2}, {y2}")
                continue

            crop_path = os.path.join(save_dir, f"{base_name}_crop_{i}.jpg")
            success = cv2.imwrite(crop_path, crop)
            if success:
                print(f"âœ… í¬ë¡­ ì €ì¥ ì„±ê³µ: {crop_path}")
            else:
                print(f"âŒ í¬ë¡­ ì €ì¥ ì‹¤íŒ¨: {crop_path}")

        except Exception as e:
            print(f"âŒ ì˜ˆì™¸ ë°œìƒ: {e} - box: {box.tolist()}")

# ì‚¬ìš© ì˜ˆì‹œ
input_folder = '/content/drive/MyDrive/Unzipped/yolo/images'
save_crop_folder = '/content/drive/MyDrive/crop'

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        image_path = os.path.join(input_folder, filename)
        yolo_detect_and_crop_debug(image_path, save_crop_folder)

print("âœ… ì „ì²´ ë””ë²„ê¹… í¬ë¡­ ì™„ë£Œ")
```

```python


# 3. YOLOv8 ëª¨ë¸ ë¡œë“œ (Google Drive ê²½ë¡œì— best.pt íŒŒì¼ ìœ„ì¹˜ ë§ê²Œ ìˆ˜ì •)
from ultralytics import YOLO
model = YOLO('/content/drive/MyDrive/yolol8_checkpoints/exp/weights/best.pt')

# 4. í¬ë¡­ ì €ì¥ í•¨ìˆ˜
import os
import cv2

def yolo_crop_save(image_path, save_dir, conf_threshold=0.25):
    os.makedirs(save_dir, exist_ok=True)
    img = cv2.imread(image_path)
    if img is None:
        print(f"âŒ ì´ë¯¸ì§€ ë¡œë“œ ì‹¤íŒ¨: {image_path}")
        return

    results = model(img, conf=conf_threshold)
    boxes = results[0].boxes

    if len(boxes) == 0:
        print(f"â— íƒì§€ëœ ë°•ìŠ¤ ì—†ìŒ: {image_path}")
        return

    base_name = os.path.splitext(os.path.basename(image_path))[0]

    for i, box in enumerate(boxes.xyxy):
        x1, y1, x2, y2 = map(int, box.tolist())
        x1 = max(0, x1)
        y1 = max(0, y1)
        x2 = min(img.shape[1], x2)
        y2 = min(img.shape[0], y2)

        crop_img = img[y1:y2, x1:x2]

        if crop_img.size == 0:
            print(f"âš ï¸ ë¹ˆ í¬ë¡­ ì˜ì—­ ë¬´ì‹œ: {x1},{y1},{x2},{y2}")
            continue

        save_path = os.path.join(save_dir, f"{base_name}_crop_{i}.jpg")
        cv2.imwrite(save_path, crop_img)
        print(f"âœ… ì €ì¥: {save_path}")

# 5. í´ë” ë‚´ ëª¨ë“  ì´ë¯¸ì§€ ì¼ê´„ ì²˜ë¦¬
def yolo_crop_save_folder(input_folder, save_dir, conf_threshold=0.25):
    for fname in os.listdir(input_folder):
        if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
            image_path = os.path.join(input_folder, fname)
            yolo_crop_save(image_path, save_dir, conf_threshold)

    print(f"âœ… ì „ì²´ í´ë” ì²˜ë¦¬ ì™„ë£Œ. í¬ë¡­ ì´ë¯¸ì§€ ì €ì¥ ìœ„ì¹˜: {save_dir}")

# 6. ì‹¤í–‰ ì˜ˆì‹œ (ë³¸ì¸ Google Drive ê²½ë¡œì— ë§ê²Œ ìˆ˜ì •)
input_folder = '/content/drive/MyDrive/Unzipped/yolo/images/train/retro'
output_folder = '/content/drive/MyDrive/crop'

yolo_crop_save_folder(input_folder, output_folder, conf_threshold=0.1)
```
